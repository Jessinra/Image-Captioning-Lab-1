# Plot Training Loss

```python
import matplotlib.pyplot as plt
import json
```

```python
training_time = "2020-03-03 17:40"
```

```python
config = open("../checkpoints/train/{}/config.txt".format(training_time), "r").read()
config = str(config.replace("'", '"').replace("True", 'true'))
config = json.loads(config)
config
```

```python
log = open("../checkpoints/train/{}/log.txt".format(training_time), "r").readlines()
log
```

```python
loss = [float(x[x.find('Loss') + 5 :].strip()) for x in log]

plt.figure(figsize=(13,13))
plt.plot(loss);
```
